### <a name='a1'></a> Import bibliotek

````
import numpy as np
import pandas as pd
````

### <a name='a2'></a> Załadowanie danych

````
def fetch_financial_data(company='AMZN'):
    """
    This function fetch stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('FB')
df.head()
````

### <a name='a3'></a> Zapis/Odczyt danych z pliku CSV

````
df.to_csv('fb.csv') #zapisanie danych do pliku csv
````

````
df_may = df[(df.index.month == 5) & (df.index.year == 2021)]
df_may
````

````
df_may.to_csv('fb_may.csv')
````

````
new_df = pd.read_csv('fb_may.csv', index_col=0)
new_df
````

````
df_may.to_excel('fb_may.xlsx')
````

````
new_df = pd.read_excel('fb_may.xlsx', index_col=0)#index_col ustawia pierwszą kolumnę jako indeks
new_df
````
