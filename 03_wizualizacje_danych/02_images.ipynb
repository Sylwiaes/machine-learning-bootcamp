````
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
````

````
img = mpimg.imread('office.jpg') #wyświetlenie obrazu jako tablicy numpy
img
````

````
img.shape #wymiar -426x640 pikseli, obiekt składa się z 3 kanałów
````

````
plt.imshow(img) #wyświetlenie obrazu
_ = plt.axis('Off') #wyłączenie osi
````

````
fig = plt.figure(figsize=(14, 12))
plt.subplot(221)
plt.imshow(mpimg.imread('office.jpg'))
plt.axis('Off')
plt.subplot(222)
plt.axis('Off')
plt.imshow(mpimg.imread('business.jpg'))
plt.subplot(223)
plt.axis('Off')
plt.imshow(mpimg.imread('children.jpg'))
plt.subplot(224)
plt.imshow(mpimg.imread('workplace.jpg'))
plt.axis('Off')
````

````
fig = plt.figure(figsize=(14, 12)) #zautomatyzowanie powyższego kodu

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']): #funkcja enumerate pozwala zwrócić indeks idx
    plt.subplot(220 + idx + 1) #dodajemy index 
    plt.imshow(mpimg.imread(img))
    plt.axis('Off')
````

````
fig = plt.figure(figsize=(14, 12))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
    plt.subplot(140 + idx + 1) #zmieniamy położenie obrazów, 1 wiersz i 4 kolumny
    plt.imshow(mpimg.imread(img))
    plt.axis('Off')
````
