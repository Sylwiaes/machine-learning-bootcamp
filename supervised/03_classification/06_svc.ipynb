### <a name='0'></a> Import bibliotek

````
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(font_scale=1.3)
np.random.seed(42)
````

### <a name='1'></a> Wygenerowanie danych

````
from sklearn.datasets import load_iris

raw_data = load_iris()
all_data = raw_data.copy()

data = all_data['data']
target = all_data['target']
feature_names = all_data['feature_names']
target_names = all_data['target_names']

df = pd.DataFrame(np.c_[data, target], columns=feature_names + ['target']) #Ograniczenie się tylko do 2 klas
# df = df[(df['target'] == 0.0) | (df['target'] == 1.0)]
df.head()
````

````
data = df.iloc[:, [2, 1]].values
target = df['target'].apply(int).values

print(f'{data[:5]}\n')
print(f'{target[:5]}')
````

### <a name='2'></a> Podział na zbiór treningowy i testowy

````
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target)
````

### <a name='3'></a> Standaryzacja cech

````
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() #Stworzenie instancji naszej klasy
scaler.fit(X_train) #Dopasowanie do danych treningowych

X_train = scaler.transform(X_train) #Transformujemy dane treningowe i testowe
X_test = scaler.transform(X_test)
````

### <a name='4'></a> Klasyfikator SVM - linear

````
from sklearn.svm import SVC

classifier = SVC(C=1.0, kernel='linear') #Parametr C kontroluje czułość modelu, domyślnie C=1; mniejsze wartości C powodują poszerzenie pasa, większe zwężenie
#Parametr kernel to jądro
classifier.fit(X_train, y_train)#Dopasowanie danych do modelu
````

````
classifier.score(X_test, y_test) #sprawdzenie jak nasz model działa na danych testowych
````

````
from mlxtend.plotting import plot_decision_regions #Granice decyzyjne dla zbioru treningowego

plt.figure(figsize=(8, 6))
plot_decision_regions(X_train, y_train, classifier)
plt.xlabel(feature_names[2])
plt.ylabel(feature_names[1])
plt.title(f'SVC: train accuracy: {classifier.score(X_train, y_train):.4f}')
plt.show()
````

````
plt.figure(figsize=(8, 6)) #Granice decyzyjne dla zbioru testowego
plot_decision_regions(X_test, y_test, classifier)
plt.xlabel(feature_names[2])
plt.ylabel(feature_names[1])
plt.title(f'SVC: test accuracy: {classifier.score(X_test, y_test):.4f}')
plt.show()
````

### <a name='5'></a> Klasyfikator SVM - rbf

````
classifier = SVC(C=1.0, kernel='rbf') #Granice decyzyjne modeli nieliniowych jako drugi sposób rozwiązania problemu

classifier.fit(X_train, y_train)
classifier.score(X_test, y_test)
````

````
plt.figure(figsize=(8, 6)) #Granice decyzyjne dla zbioru treningowego
plot_decision_regions(X_train, y_train, classifier)
plt.xlabel(feature_names[2])
plt.ylabel(feature_names[1])
plt.title(f'SVC: train accuracy: {classifier.score(X_train, y_train):.4f}')
plt.show()
````

````
plt.figure(figsize=(8, 6)) #Granice decyzyjne dla zbioru testowego
plot_decision_regions(X_test, y_test, classifier)
plt.xlabel(feature_names[2])
plt.ylabel(feature_names[1])
plt.title(f'SVC: test accuracy: {classifier.score(X_test, y_test):.4f}')
plt.show()
````
