### <a name='a4'></a> K-nearest Neighbour Algorithm - Algorytm K-najbliższych sąsiadów

Podstawą działania algorytmu jest:
* znalezienie z góry określonej liczby próbek treningowych znajdujących się najbliżej naszej obserwacji 
* przewidzenie na ich podstawie etykiety

Liczba sąsiadów jest określana przez użytkownika. Odległości zwykle kalkuluje sie przy pomocy metryki euklidesowej. 

````
from sklearn.neighbors import KNeighborsClassifier #Zaimportowanie klasy k-najbliższych sąsiadów

classifier = KNeighborsClassifier(n_neighbors=5) #bierzemy pod uwagę 5 najbliższych sasiadów; nie rozgraniczamy danych na zbiór testowy i treningowy, bo 
#naszych danych jest niewiele i uczymy się jak nasz model działa
classifier.fit(data, target)
````

### <a name='a5'></a> Wykres granic decyzyjnych

````
x_min, x_max = data[:, 0].min() - 0.5, data[:, 0].max() + 0.5 #Pierwszy krok to wyznaczenie wartości skrajnych; bierzemy dla danych z pierwszej zmiennej min,
#odejmujemy 0,5 w celu dodania brzegów na wykresie, aby wykres był bardziej czytelny
y_min, y_max = data[:, 1].min() - 0.5, data[:, 1].max() + 0.5

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01)) #Utworzenie siatki, dzięki której możemy użyć metody predict i przewidzieć
#dla każdego puntku teh siatki konkretną klasę
mesh = np.c_[xx.ravel(), yy.ravel()]
Z = classifier.predict(mesh)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 8))
plt.pcolormesh(xx, yy, Z, cmap='gnuplot', alpha=0.1) #Funkcja pozwala na wyświetlenie siatki i granic decyzyjnych
plt.scatter(data[:, 0], data[:, 1], c=target, cmap='gnuplot', edgecolors='r')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title('3-class classification k=5')
plt.show()
````

````
 def plot_decision_boundries(n_neighbors=1): #Budowanie funkcji, która upraszcza budowanie granic decyzyjnych i tworzenia modelu; domyślnie ustawiamy k=1 
     
    classifier = KNeighborsClassifier(n_neighbors=n_neighbors) #Budujemy klasyfikator
    classifier.fit(data, target) #Dopasowujemy dane do klasyfikatora

    x_min, x_max = data[:, 0].min() - 0.5, data[:, 0].max() + 0.5
    y_min, y_max = data[:, 1].min() - 0.5, data[:, 1].max() + 0.5

    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    mesh = np.c_[xx.ravel(), yy.ravel()]
    Z = classifier.predict(mesh)
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(10, 8))
    plt.pcolormesh(xx, yy, Z, cmap='gnuplot', alpha=0.1)
    plt.scatter(data[:, 0], data[:, 1], c=target, cmap='gnuplot', edgecolors='r')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(f'3-class classification k={n_neighbors}')
    plt.show()

plot_decision_boundries()
````

````
plot_decision_boundries(2) #Wyświetlenie 2 najblizszych sasiadów
````

````
plot_decision_boundries(50)
````

````
plt.figure(figsize=(12, 12)) #Porównanie kilak różnych wartości dla parametru k przy pomocy biblioteki matplotlib

for i in range(1, 7):
    plt.subplot(3, 2, i)

    clf = KNeighborsClassifier(n_neighbors=i)
    clf.fit(data, target)

    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    mesh = np.c_[xx.ravel(), yy.ravel()]
    Z = clf.predict(mesh)
    Z = Z.reshape(xx.shape)

    plt.pcolormesh(xx, yy, Z, cmap='gnuplot', alpha=0.1)
    plt.scatter(data[:, 0], data[:, 1], c=target, cmap='gnuplot', edgecolors='r')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(f'3-class classification k={i}')

plt.show()
````
